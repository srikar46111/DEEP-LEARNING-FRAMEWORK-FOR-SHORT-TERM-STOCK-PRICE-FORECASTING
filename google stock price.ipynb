```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from datetime import timedelta

# -------------------------
# 1. Load Google Stock Data
# -------------------------
df = yf.download("GOOG", start="2010-01-01", end="2025-12-31")
print("Dataset shape:", df.shape)
print(df.tail())

# Use 'Open' price
dataset = df[['Open']].values

# -------------------------
# 2. Feature Scaling
# -------------------------
sc = MinMaxScaler(feature_range=(0,1))
dataset_scaled = sc.fit_transform(dataset)

# -------------------------
# 3. Train-Test Split
# -------------------------
training_size = int(len(dataset_scaled) * 0.8)
train_data = dataset_scaled[:training_size]
test_data  = dataset_scaled[training_size-60:]

# -------------------------
# 4. Prepare Training Data
# -------------------------
X_train, y_train = [], []
for i in range(60, len(train_data)):
    X_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

# -------------------------
# 5. Build LSTM Model
# -------------------------
regressor = Sequential()
regressor.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units=50))
regressor.add(Dropout(0.2))
regressor.add(Dense(units=1))

regressor.compile(optimizer='adam', loss='mean_squared_error')
regressor.fit(X_train, y_train, epochs=50, batch_size=32)

# -------------------------
# 6. Prepare Test Data
# -------------------------
X_test, y_test = [], []
for i in range(60, len(test_data)):
    X_test.append(test_data[i-60:i, 0])
    y_test.append(test_data[i, 0])

X_test, y_test = np.array(X_test), np.array(y_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# -------------------------
# 7. Predictions on Test Set
# -------------------------
predicted_stock_price = regressor.predict(X_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
real_stock_price = sc.inverse_transform(y_test.reshape(-1,1))

# -------------------------
# 8. Visualization (Test Results)
# -------------------------
plt.figure(figsize=(12,6))
plt.plot(real_stock_price, color='red', label='Real Google Stock Price')
plt.plot(predicted_stock_price, color='blue', label='Predicted Google Stock Price')
plt.title('Google Stock Price Prediction (Test Data)')
plt.xlabel('Time')
plt.ylabel('Stock Price (USD)')
plt.legend()
plt.show()

# -------------------------
# 9. Future Prediction (Next 30 Days)
# -------------------------
last_60_days = dataset_scaled[-60:]  # last 60 days from full dataset
future_input = last_60_days.reshape(1, 60, 1)  # Reshape to match model input shape

future_predictions = []
n_future_days = 30

for _ in range(n_future_days):
    # Predict next day's price
    pred = regressor.predict(future_input)
    future_predictions.append(pred[0,0])

    # Update future_input: remove the first day and add the prediction
    future_input = np.append(future_input[:, 1:, :], pred.reshape(1, 1, 1), axis=1)


future_predictions = sc.inverse_transform(np.array(future_predictions).reshape(-1,1))

# -------------------------
# 10. Create Dates for Future Forecast
# -------------------------
last_date = df.index[-1]   # last available trading day
future_dates = [last_date + timedelta(days=i+1) for i in range(n_future_days)]

forecast_df = pd.DataFrame({
    "Date": future_dates,
    "Predicted_Open": future_predictions.flatten()
})

# -------------------------
# 11. Save to CSV
# -------------------------
forecast_df.to_csv("Google_Future_30_Days.csv", index=False)
print("✅ Future forecast saved as Google_Future_30_Days.csv")

# -------------------------
# 12. Visualization (Future Forecast)
# -------------------------
plt.figure(figsize=(12,6))
plt.plot(real_stock_price, color='red', label='Real Google Stock Price (Test)')
plt.plot(predicted_stock_price, color='blue', label='Predicted (Test)')
plt.plot(range(len(real_stock_price), len(real_stock_price)+n_future_days),
         future_predictions, color='green', linestyle='dashed', label='Future Forecast (30 days)')
plt.title('Google Stock Price - Future 30 Days Forecast')
plt.xlabel('Time')
plt.ylabel('Stock Price (USD)')
plt.legend()
plt.show()
```
